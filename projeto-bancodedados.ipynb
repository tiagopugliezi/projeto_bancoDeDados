```python
#Questionamento #3 
#Qual o mês com a maior receita de vendas da Van Arsdel?

sql = """select sum(revenue) as total, substring(date from 4 for 2) as mes 
        FROM vendas 
        inner join produtos on vendas.productid = produtos.productid 
        inner join fabricantes on produtos.manufacturerid = fabricantes.manufacturerid 
        where manufacturer = 'VanArsdel'
        group by mes 
        order by total desc 
        limit 1
"""
df = pd.read_sql(sql, connection)
df
```

```python
#Questionamento #4 
#Qual a média de receita do setor por ano?

sql = """select AVG(revenue), substring(date from 7 for 4) as ano 
        FROM vendas 
        inner join produtos on vendas.productid = produtos.productid
        group by ano
        order by ano desc
"""
df = pd.read_sql(sql, connection)
df        
```

```python
#Questionamento #5 
#Qual o segmento mais lucrativo para a Van Arsdel?
sql = """select sum(revenue) as total, segment 
        FROM vendas 
        inner join produtos on vendas.productid = produtos.productid 
        inner join fabricantes on produtos.manufacturerid = fabricantes.manufacturerid 
        where manufacturer = 'VanArsdel'
        group by segment 
        order by total desc 
        limit 1
"""
df = pd.read_sql(sql, connection)
df
```
